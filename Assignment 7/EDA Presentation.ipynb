**Data Cleaning**

Converting Data Type

```python
df["order_date"] = pd.to_datetime(df["order_date"],  dayfirst=True , format = "mixed" )
df["ship_date"] = pd.to_datetime(df["ship_date"],  dayfirst=True , format = "mixed" )

```

```python
df
```

```python
df["sales"] = df["sales"].str.replace(",",'').str.strip()
df['sales'] = df['sales'].astype(int)
```

```python
df.columns = df.columns.str.replace('ï»¿', '').str.strip()
```

```python
g = df.groupby('product_name')['quantity'].sum().sort_values(ascending=False).head(10)
plt.barh(g.index , g.values )
plt.ylabel("Product Name")
plt.xlabel("Quantity")
plt.title('Top ten selling products')
plt.show()
```

```python
import seaborn as sns
sns.relplot(data = df ,x = 'profit' , y = 'sales' , color = 'red'  )
plt.title('Relationship Between Sales and Profit' , weight = 'bold')
plt.xlabel('Profit' , fontsize = 12)
plt.ylabel('Sales' , fontsize = 12)
plt.show()
```

```python
sns.relplot(data = df ,x = 'discount' , y = 'profit' , color = 'blue'  )
plt.title('Relationship Between Discount and Profit' , weight = 'bold')
plt.xlabel('Discount' , fontsize = 12)
plt.ylabel('Profit' , fontsize = 12)
plt.show()
```

```python
df['year'] = df['order_date'].dt.year
```

```python
df
```

```python
sales_by_year = df.groupby('year')['sales'].sum()
```

```python
sales_by_year
```

```python
plt.bar(sales_by_year.index , sales_by_year.values,   color= 'red' )
plt.xlabel("Years")
plt.ylabel('Sales(in Million)')
plt.xticks(sales_by_year.index)
plt.show()
```

```python
b = df.groupby('customer_name')['sales'].sum().sort_values(ascending=False).head(10)
plt.barh(b.index , b.values , color = 'orange')
plt.ylabel("Customer Name")
plt.xlabel("Sales")
plt.title('Top ten Customers')
plt.show()
```

```python
b = df.groupby('country')['sales'].sum().sort_values(ascending=False).head(10)
plt.barh(b.index , b.values,color = 'red')
plt.ylabel("Customer Name")
plt.xlabel("Sales(in Million)")
plt.title('Top ten countries')
plt.show()
```
