```python
c = a.reshape(2,6)
c
```

**Flatten**

```python
c1 = c.flatten()
c1
```

```python
c1[6] = 140
c1
```

```python
c
```

**No changes occur in original array**

**Ravel**

```python
c2 = c.ravel()
c2
```

```python
c2[8] = 50
c2
```

```python
c
```
